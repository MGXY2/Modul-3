```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
```

```python
x = [0,1,2]
y=[1,3,2]
f= interp1d(x,y)
y_hat=f(1.5)
print(y_hat)
```

```python
plt.figure(figsize=(10,8))
plt.plot(x,y,'-ob')
plt.plot(1.5,y_hat,'ro')
plt.title('linear interpolation at x = 1.5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def divided_diff(x, y):
    n = len(y)
    coef = np.zeros([n, n])  # Membuat tabel kosong untuk divided differences
    coef[:, 0] = y  # Kolom pertama berisi nilai y

    for j in range(1, n):  # Loop untuk mengisi kolom-kolom berikutnya
        for i in range(n - j):
            coef[i][j] = (coef[i + 1][j - 1] - coef[i][j - 1]) / (x[i + j] - x[i])

    return coef
def newton_poly(coef, x_data, x):
    n = len(x_data) - 1
    p = coef[n]

    for k in range(1, n + 1):
        p = coef[n - k] + (x - x_data[n - k]) * p

    return p
```

```python
x=np.array([-5,-1,0,2])
y=np.array([-2,6,1,3])

a_s=divided_diff(x,y)[0,:]

x_new = np.arange(-5,2.1,.1)
y_new = newton_poly(a_s,x,x_new)

plt.figure(figsize=(12,8))
plt.plot(x,y,'bo')
plt.plot(x_new,y_new)
```

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

x=[0,1,2]
y=[1,3,2]

f = CubicSpline(x,y,bc_type='natural')

x_new=np.linspace(0,2,100)
y_new=f(x_new)

plt.figure(figsize=(10,8))
plt.plot(x_new,y_new,'b')
plt.plot(x,y,'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

jarak = np.array([0, 5, 10])
kecepatan = np.array([10, 15, 20])

f = interp1d(jarak, kecepatan)

jarak_baru = np.linspace(0, 10, 100)
kecepatan_baru = f(jarak_baru)

plt.plot(jarak, kecepatan, 'o', label='Data Asli')
plt.plot(jarak_baru, kecepatan_baru, '-', label='Interpolasi Linear')
plt.plot(7.2, f(7.2), 'ro', label='Kecepatan pada 7.2m')

plt.title('Interpolasi Linear Kecepatan Lari Ben Johnson')
plt.xlabel('Jarak (meter)')
plt.ylabel('Kecepatan (meter/detik)')
plt.legend()
plt.grid(True)
plt.show()
```
